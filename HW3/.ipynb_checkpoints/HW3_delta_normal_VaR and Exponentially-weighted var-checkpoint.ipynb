```python
import numpy as np
from tqdm import tqdm
import pandas as pd
```

```python
assets = ['XLB', 'XLE', 'XLF', 'XLP', 'XLV', 'XLY']

df = pd.read_csv('C:/Users/kamin/OneDrive/Documents/min/mfe/uiuc/Spring2024/Risk Management/Week4/data.csv', 
                 parse_dates=['Date'], index_col='Date')
df = df.iloc[-500:][assets]
last_prices = np.array([77.68000, 87.47000, 34.20000, 74.55000, 135.85001, 129.16000])

holding_volumn = np.array([30000, 30000, 60000, 30000, 15000, 10000])

asset_val = holding_volumn * last_prices
```

```python
lambd = 0.94
lambd_weight = [lambd**(i) for i in range(500)][::-1]

df['lambda weight'] = lambd_weight
df
```

```python
cov = []

for i in assets:
    for j in assets:
        cov.append((df[i] * df[j] * df['lambda weight']).sum() * (1-lambd))
cov = pd.DataFrame(np.reshape(cov, (len(assets), len(assets))), index=assets, columns=assets)
cov
```

```python
asset_value_vector = asset_val.reshape((1,-1))
0 - 1.645*np.sqrt(asset_value_vector@cov.values@asset_value_vector.T)#Assume: zero expected returns
```
