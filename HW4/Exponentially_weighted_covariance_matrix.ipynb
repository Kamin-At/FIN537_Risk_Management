```python
import pandas as pd
import numpy as np
```

```python
lambd = 0.94

assets = ['log_ret_xle', 'log_ret_xlf', 'log_ret_xly', 'log_vol_xle', 'log_vol_xlf', 'log_vol_xly']

df = pd.read_csv("returns.csv")[["Date"] + assets]
df['lambda weight'] = lambd ** np.arange(499, -1,-1)
df
```

```python
cov = []

for i in assets:
    for j in assets:
        cov.append((df[i] * df[j] * df['lambda weight']).sum() * (1-lambd))
cov = pd.DataFrame(np.reshape(cov, (len(assets), len(assets))), index=assets, columns=assets)
cov
```
